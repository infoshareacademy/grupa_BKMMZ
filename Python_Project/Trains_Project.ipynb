Purpose: to analyze train delays in France
Stakeholder: a railroad manager who needs to identify the reasons for delays on selected rail lines

Year i Month: Pozwalają na analizę trendów opóźnień w różnych latach i miesiącach.

Departure station i Arrival station: Pozwalają zidentyfikować, które stacje są bardziej narażone na opóźnienia.

Average travel time (min): Pomaga zrozumieć, czy opóźnienia wpływają na czas podróży.

Number of expected circulations, Number of cancelled trains, Number of late trains at departure: Dają informacje o ogólnej wydajności linii kolejowych.

% trains late due to różne przyczyny: Pozwala zidentyfikować główne przyczyny opóźnień, takie jak warunki atmosferyczne, problemy infrastrukturalne itp.

Delay due to różne przyczyny: Podobnie jak powyżej, pomaga zrozumieć, które czynniki są głównymi przyczynami opóźnień.

Comment (optional) delays at departure i Comment (optional) delays on arrival: Mogą zawierać dodatkowe informacje na temat specyficznych sytuacji związanych z opóźnieniami.

Period: Pomaga w analizie opóźnień w określonym przedziale czasowym.
